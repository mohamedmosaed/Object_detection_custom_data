```python
from os import listdir
from torch import nn
import matplotlib.pyplot as plt
from detecto.utils import read_image
from detecto.core import Dataset
from detecto.visualize import show_labeled_image
from detecto.core import DataLoader, Model
```

### Training DataSet

```python
based_path = r"C:\Users\mosae\USD Workes\AAI 521 Intro to Computer Vision\final_ptoject_1\train"
train_dataset = Dataset(based_path)
```

```python
image, targets = train_dataset[9]
show_labeled_image(image, targets['boxes'], targets['labels'])
```

### validation Dataset

```python
based_path = r"C:\Users\mosae\USD Workes\AAI 521 Intro to Computer Vision\final_ptoject_1\test"
val_dataset = Dataset(based_path)
```

## train the model

```python
# specifiy the labels
labels = ['LILY']
model = Model(labels)
```

```python
print(model)
```

```python
# fit the model
losses = model.fit(train_dataset, val_dataset, epochs=15)
```

```python
plt.plot(losses)
plt.show()
```

## visualizing the model prediction 

```python
image = read_image(r'C:\Users\mosae\USD Workes\AAI 521 Intro to Computer Vision\final_ptoject_1\test\RIUD2947.JPG')
plt.imshow(image)
plt.show()
```

```python
labels, boxes, score = model.predict(image)
```

```python
print("labels", labels)
print("score", score)
print("boxes", boxes)
```
